```python
flight = df3.drop(columns=['Unnamed: 0', 'flight'])
```

```python
flight_dummies = pd.get_dummies(flight)
flight_dummies.head()
```

```python
features = flight_dummies.drop(columns='price')
target = flight_dummies['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(features, target, random_state=13)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
lr.score(X_test, Y_test)
```
